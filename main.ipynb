importing the python library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

data for movie genre what people like most

```python
data = {
    "Genre": [
        "Romance", "Comedy", "Horror", "Action", "Science Fiction", 
        "Drama", "Animated", "Documentary", "Fantasy", "Mystery"
    ],
    "Average Rating": [
        4.5, 4.0, 3.8, 4.2, 4.3, 
        4.6, 4.4, 4.1, 4.3, 4.0
    ],
    "Ratings Count": [
        1200, 1500, 900, 1300, 1100, 
        1400, 1000, 600, 950, 700
    ]
}
```

```python
# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Display basic information and statistics
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Plotting the data
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for Ratings Count
bars = ax1.bar(df["Genre"], df["Ratings Count"], color='lightblue', alpha=0.6, label='Ratings Count')

# Line plot for Average Rating
ax2 = ax1.twinx()
line = ax2.plot(df["Genre"], df["Average Rating"], color='red', marker='o', linestyle='-', label='Average Rating')

# Adding titles and labels
ax1.set_xlabel('Movie Genre', fontsize=12)
ax1.set_ylabel('Ratings Count', fontsize=12, color='blue')
ax2.set_ylabel('Average Rating', fontsize=12, color='red')
ax1.set_title('Movie Genre Popularity', fontsize=14)

# Adding a legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
```

```python
# Displaying the plot
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Determining the most loved genre
df['Total Score'] = df['Average Rating'] * df['Ratings Count']
most_loved_genre = df.loc[df['Total Score'].idxmax()]

print(f"The most loved genre is {most_loved_genre['Genre']} with an average rating of {most_loved_genre['Average Rating']} and {most_loved_genre['Ratings Count']} ratings.")
```
